```python
yearlist = [str(i) for i in [*range(2014, 2020)]]
```

```python
typeofcrimelist = ['Assault', 'AutoTheft', 'BreakandEnter',  'Robbery', 'TheftOver']
```

```python
crimeyear = []
for crime in typeofcrimelist:
    for year in yearlist:
        crimeyear.append(crime + '_' + year)
        
```

```python
crimeyear
```

```python
for crime in typeofcrimelist:
    df[crime + '_CHG_2014_2015']= (df[crime+'_2015']- df[crime+'_2014'] )/df[crime+'_2014'] 

for crime in typeofcrimelist:
    df[crime + '_CHG_2015_2016']= (df[crime+'_2016']- df[crime+'_2015'] )/df[crime+'_2015'] 

for crime in typeofcrimelist:
    df[crime + '_CHG_2016_2017']= (df[crime+'_2017']- df[crime+'_2016'] )/df[crime+'_2016'] 

for crime in typeofcrimelist:
    df[crime + '_CHG_2017_2018']= (df[crime+'_2018']- df[crime+'_2017'] )/df[crime+'_2017'] 

for crime in typeofcrimelist:
    df[crime + '_CHG_2018_2019']= (df[crime+'_2019']- df[crime+'_2018'] )/df[crime+'_2018'] 
```

```python
print(df.isna().sum())
```

```python
new_df = df.drop(columns=crimeyear)
```
